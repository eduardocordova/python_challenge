```python
with open('-B166OkVBPJR/2019-09-02_19-27-12_UTC_comments.json','r') as f:
    data = json.loads(f.read())
    
df_nested_list=pd.json_normalize(data)
df_nested_list
```

```python
with open('-B166OkVBPJR/2019-09-02_19-27-12_UTC_comments.json','r') as f:
    data = json.loads(f.read())
    
df_nested_list_answers=pd.json_normalize(data, record_path=['answers'], record_prefix='a_', meta=['id'])
```

```python
df_nested_list_answers
```

```python
comments = df_nested_list.rename(columns={'id': 'IdFatherComment',
                                          'text': 'Caption',
                                          'created_at': 'Date',
                                          'likes_count': 'likesComment',
                                          'owner.username': 'Username'
                                         })
```

```python
comments
```

```python
answers = df_nested_list_answers.rename(columns={'a_id': 'IdChildComment',
                                          'a_text': 'Caption',
                                          'a_created_at': 'Date',
                                          'a_likes_count': 'likesComment',
                                          'a_owner.username': 'Username',
                                                 'id': 'IdFatherComment'
                                         })
#answers['IdChildComment'].astype(str)

answers['IdChildComment'] = answers['IdChildComment'].apply(str)
```

```python
answers
```

```python
total_comments = pd.concat([comments, answers], sort=False)

del total_comments['answers']
del total_comments['owner.id']
del total_comments['owner.is_verified']
del total_comments['owner.profile_pic_url']
del total_comments['a_owner.id']
del total_comments['a_owner.is_verified']
del total_comments['a_owner.profile_pic_url']
```

```python
total_comments.insert(0, 'Post', range(0, len(total_comments)))
```
